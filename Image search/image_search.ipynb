## 1 Исходные данные

### 1.1 Описание проекта

Сервис фотохостинга для профессиональных фотографов занимается экспериментом по разработке поиска референсных фотографий для посетителей.\
Суть поиска заключается в следующем: пользователь сервиса вводит описание нужной сцены, а сервис выводит несколько фотографий с такой же или похожей сценой.
Чтобы эксперимент получил право на жизнь, нужно защитить его перед руководителем компании.\
Для защиты необходимо презентовать так называемый PoC (Proof of Concept, Проверка концепции) — продемонстрировать, что такой проект практически осуществим.

### 1.2 Описание данных

В файле `train_dataset.csv` находится информация, необходимая для обучения: имя файла изображения, идентификатор описания и текст описания. Для одной картинки может быть доступно до 5 описаний. Идентификатор описания имеет формат `<имя файла изображения>#<порядковый номер описания>`.

В папке `train_images` содержатся изображения для тренировки модели.
В файле `CrowdAnnotations.tsv`  — данные по соответствию изображения и описания, полученные с помощью краудсорсинга. Номера колонок и соответствующий тип данных:
- Имя файла изображения.
- Идентификатор описания.
- Доля людей, подтвердивших, что описание соответствует изображению.
- Количество человек, подтвердивших, что описание соответствует изображению.
- Количество человек, подтвердивших, что описание не соответствует изображению.

В файле `ExpertAnnotations.tsv`  — данные по соответствию изображения и описания, полученные в результате опроса экспертов. Номера колонок и соответствующий тип данных:
- Имя файла изображения.
- Идентификатор описания.
- 3, 4, 5 — оценки трёх экспертов.

Эксперты ставят оценки по шкале от 1 до 4, где 1 — изображение и запрос совершенно не соответствуют друг другу, 2 — запрос содержит элементы описания изображения, но в целом запрос тексту не соответствует, 3 — запрос и текст соответствуют с точностью до некоторых деталей, 4 — запрос и текст соответствуют полностью.

В файле `test_queries.csv` находится информация, необходимая для тестирования: идентификатор запроса, текст запроса и релевантное изображение. Для одной картинки может быть доступно до 5 описаний. Идентификатор описания имеет формат `<имя файла изображения>#<порядковый номер описания>`.

В папке `test_images` содержатся изображения для тестирования модели.

### 4.1 Агрегация оценок экспертов и краудсорсинга

Для того, чтобы сделать агрегацию оценок экспертов и краудсорсинга необходимо, чтобы их оценки были в одном масштабе. Для этого в датасете с оценками экспертов создадим стоблец со средним значением оценок и отмасштабируем его значения:

```python
#df_expert_annots['ratio'] = df_expert_annots[['score_1', 'score_2', 'score_3']].mean(axis=1)
df_expert_annots.loc[:, 'score_1':'score_3'] = df_expert_annots.loc[:, 'score_1':'score_3'] - 1
df_expert_annots['ratio_expert'] = df_expert_annots.loc[:, 'score_1':'score_3'].mean(axis=1)
df_expert_annots.sample(5)
```

Теперь объединим данные двух датасетов:

```python
df_crowd_annots = df_crowd_annots.rename(columns={'ratio' : 'ratio_crowd'})

df_annots_agg = df_crowd_annots[['image', 'query_id', 'ratio_crowd']].merge(
    df_expert_annots[['image', 'query_id', 'ratio_expert']], how='outer', on=['query_id', 'image']
    )
print(df_annots_agg.shape)
df_annots_agg.sample(5)
```

Рассчитаем агрегированную оценку, в которой оценка эксперта принимается с коэффициентом 0.6, а крауда — с коэффициентом 0.4:

```python
df_annots_agg['ratio_agg'] = np.nan

def func_ratio_agg(row):
    if not pd.isnull(row['ratio_expert']) and not pd.isnull(row['ratio_crowd']):
      row['ratio_agg'] = 0.6*row['ratio_expert'] + 0.4*row['ratio_crowd']
    elif not pd.isnull(row['ratio_expert']):
      row['ratio_agg'] = row['ratio_expert']
    elif not pd.isnull(row['ratio_crowd']):
      row['ratio_agg'] = row['ratio_crowd']
    return row

df_annots_agg = df_annots_agg.apply(func_ratio_agg, axis=1)
```

```python
df_annots_agg.sample(5)
```

```python
df_train_cleared = df_annots_agg[['image', 'query_id', 'ratio_agg']].merge(
    df_train[['query_text', 'query_id']].drop_duplicates(), how='inner', on='query_id'
)
print(df_train_cleared.shape)
df_train_cleared.sample(5)
```

### 4.2 Лемматизация и удаление запрещенного контента

Загрузим лемматизатор spaCy:

```python
lemmatizer = spacy.load("en_core_web_sm")
```

Далее проведем предобработку текстов - токенизацию и лемматизацию, для этого создадим функцию:

```python
@lru_cache(maxsize=12)
def func_lemm_text(text):
    text = re.sub(r"[^a-zA-Z']", ' ', text)
    text = ' '.join(text.lower().split())
    doc = lemmatizer(text)
    clear_text = " ".join([token.lemma_ for token in doc])
    return clear_text
```

```python
df_train_cleared['clear_query_text'] = df_train_cleared['query_text'].progress_apply(lambda x: func_lemm_text(x))

#Wall time - 4 mins
```

Приведем векторы признаков к одному размеру путем добавления нулей к более коротким векторам:

```python
max_len = 0
for i in tokenized.values:
    if len(i) > max_len:
        max_len = len(i)

padded = np.array([i + [0]*(max_len-len(i)) for i in tokenized.values])
```

Поскольку эти нули не несут значимой информации, отбросим эти токены и «создадим маску» для действительно важных токенов:

```python
attention_mask = np.where(padded != 0, 1, 0)
print(attention_mask.shape)
```

Создадим эмбеддинги для текстов, ограничив размер батча 50 объектами:

```python
batch_size = 50
text_embeddings = []
for i in tqdm(range(padded.shape[0] // batch_size + 1)):
        batch = torch.LongTensor(padded[batch_size*i:batch_size*(i+1)]) 
        attention_mask_batch = torch.LongTensor(attention_mask[batch_size*i:batch_size*(i+1)])
        
        with torch.no_grad():
            batch_embeddings = model_bert(batch.to(device), attention_mask=attention_mask_batch.to(device))
        
        text_embeddings.append(batch_embeddings[0][:,0,:].cpu().numpy())

#Wall time colab cpu ~1 hour 5 mins
#Wall time colab gpu ~5 mins
#Wall time yxds cpu ~10 mins
```

Запишем все эмбеддинги в матрицу признаков:

```python
features_bert = np.concatenate(text_embeddings)
features_bert.shape
```

### 4.4 Векторизация изображений

Для векторизации изображений используем сверточную нейросеть ResNet-18, предобученную на датасете ImageNet, с замороженными весами:

```python
resnet = models.resnet18(pretrained=True) #загружаем претренированную модель
# замораживаем веса
for param in resnet.parameters():
    param.requires_grad_(False)

modules = list(resnet.children())[:-1] #берем все слои, кроме последнего
#modules.append(nn.Flatten())
resnet = nn.Sequential(*modules)
resnet.eval() # переводим модель в режим предсказания
resnet = resnet.to(device)
```

Далее приведем изображения в нужный формат: изменим размер, вырежем центр и нормализуем:

```python
norm = transforms.Normalize(
    mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]
    )
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    norm,
])
```

Для чтения изображений и получения эмбеддингов, создадим функцию:

```python
def func_img_embeddings(model, base_path, df):
    img_embeddings = []
    cache = {}
    
    for path in tqdm(df['image']):
        output_tensor = cache.get(path)
        if output_tensor is None:
            img = Image.open(f'{base_path}/{path}').convert('RGB') #lib PIL
            image_tensor = preprocess(img)
            image_tensor = image_tensor.unsqueeze(0).to(device)
            output_tensor = resnet(image_tensor).flatten()
            cache[path] = output_tensor

        img_embeddings.append(output_tensor)
    
    return torch.stack(img_embeddings).cpu().detach().numpy()
```

```python
features_resnet = func_img_embeddings(resnet, f'{PATH}/train_images', df_train_filtered)
features_resnet.shape
#Wall time 35 s
```

### 5.2 Полносвязная нейронная сеть

Создадим полносвязную нейронную сеть:

```python
def func_create_model(input_shape):
    optimizer = Adam(learning_rate=1e-5)
    model = Sequential()
    model.add(Dense(1024, activation='relu', input_shape=input_shape))
    model.add(Dropout(0.3))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(1, activation='linear'))
    
    model.build()
    model.compile(
        optimizer=optimizer, 
        loss='mse'
    )
    print(model.summary())
    
    return model
```

####  5.2.2 Полносвязная нейронная сеть в Pytorch

Создадим тензоры признаков и целевого признака:

```python
X_train_tensor = torch.FloatTensor(X_train.values)
X_valid_tensor = torch.FloatTensor(X_valid.values)
y_train_tensor = torch.FloatTensor(y_train.values)
y_valid_tensor  = torch.FloatTensor(y_valid.values)
```

Выведем размерность тензоров:

```python
print('Размеры получившихся выборок:\n '
      'Признаки, обучающая выборка:', X_train_tensor.shape, '\n',
      'Признаки, валидационная выборка:', X_valid_tensor.shape, '\n',
      'Целевой признак, обучающая выборка:', y_train_tensor.shape, '\n',
      'Целевой признак, валидационная выборка:', y_valid_tensor.shape)
```

Чтобы не создавать самим батчи, воспользуемся Dataset и Dataloader:

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
valid_dataset = TensorDataset(X_valid_tensor, y_valid_tensor)

train_dataloader = DataLoader(train_dataset, batch_size=128, shuffle=True)
valid_dataloader = DataLoader(valid_dataset, batch_size=128)
```

Рассмотрим модель с двумя скрытыми слоями:

```python
class Net2(nn.Module):
    
    def __init__(
        self,
        n_in_neurons,
        n_hidden_neurons_1,
        n_hidden_neurons_2,
        #n_hidden_neurons_3,
        n_out_neurons,
        activation,
        dropout
    ):
        super().__init__()
        self.fc1 = nn.Linear(n_in_neurons, n_hidden_neurons_1)
        self.act1 = activation
        self.dp = nn.Dropout(dropout)
        self.fc2 = nn.Linear(n_hidden_neurons_1, n_hidden_neurons_2)
        self.act2 = nn.ReLU()
        #self.fc3 = nn.Linear(n_hidden_neurons_1, n_hidden_neurons_3)
        #self.act3 = nn.ReLU()
        self.fc4 = nn.Linear(n_hidden_neurons_2, n_out_neurons)
        
        nn.init.kaiming_normal_(self.fc1.weight, nonlinearity='relu')
        nn.init.normal_(self.fc1.bias, mean=0, std=(2/n_in_neurons)**0.5)
        nn.init.kaiming_normal_(self.fc2.weight, nonlinearity='relu')
        nn.init.normal_(self.fc2.bias, mean=0, std=(2/n_hidden_neurons_1)**0.5)
        #nn.init.kaiming_normal_(self.fc3.weight, nonlinearity='relu')
        #nn.init.normal_(self.fc3.bias, mean=0, std=(2/n_hidden_neurons_2)**0.5)
        
    def forward(self, x):
        x = self.fc1(x)
        x = self.dp(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.act2(x)
        #x = self.fc3(x)
        #x = self.act3(x)
        x = self.fc4(x)
        return x
```

Обучим модель:

```python
torch.manual_seed(SEED)
np.random.seed(SEED)

best_epoch = None
best_rmse = 99999
rmse_train = 99999
epochs= 80


net = Net2(X_train_tensor.shape[1], 512, 256, 1, nn.ELU(), 0).to(device)
loss = nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', threshold=0.001)

loss_tr = []
loss_val = []
rmse_tr = []
rmse_val = []

for epoch in tqdm(range(epochs)):
    if epoch % 10 == 0 or epoch == (epochs - 1):
        print('epoch', epoch)
    net.train()
    avg_loss = 0
    avg_loss_val = 0
    avg_rmse = 0
    avg_rmse_val = 0
    
    for X_batch, Y_batch in train_dataloader:
        X_batch = X_batch.to(device)
        Y_batch = Y_batch.to(device)
        optimizer.zero_grad()
        outp = net(X_batch).flatten(start_dim=-2)
        loss_value = loss(outp, Y_batch)
        loss_value.backward()
        optimizer.step()
        avg_loss += loss_value / len(train_dataloader)
        avg_rmse += mse(Y_batch.cpu(),
                                       outp.cpu().detach().numpy(),
                                       squared=False) / len(train_dataloader)
    scheduler.step(avg_loss)
    loss_tr.append(avg_loss.cpu().detach())
    rmse_tr.append(avg_rmse)
    
    if epoch % 10 == 0 or epoch == (epochs - 1):
        print('loss: %f' % avg_loss)
        print('rmse: %f' % avg_rmse)
            
    net.eval() 
    with torch.no_grad():
        for X_batch, Y_batch in valid_dataloader:
            X_batch = X_batch.to(device)
            Y_batch = Y_batch.to(device)
            preds_val = net(X_batch).flatten(start_dim=-2)
            loss_value = loss(preds_val, Y_batch)
            avg_loss_val += loss_value / len(valid_dataloader)
            avg_rmse_val += mse(Y_batch.cpu(),
                                               preds_val.cpu().numpy(),
                                               squared=False) / len(valid_dataloader)
        loss_val.append(avg_loss_val.cpu().detach())
        rmse_val.append(avg_rmse_val)
            
        if epoch % 10 == 0 or epoch == (epochs - 1):
            print('loss_val: %f' % avg_loss_val)
            print('rmse_val: %f' % avg_rmse_val)
        if avg_rmse_val < best_rmse:
            best_rmse = avg_rmse_val
            best_epoch = epoch
            rmse_train = avg_rmse

    torch.cuda.empty_cache()
```

Выведем наилучшие параметры модели - эпоху и метрику RMSE:

```python
print(f'Лучшая эпоха: {best_epoch}', 
      f'Метрика RMSE на валидационной выборке: {best_rmse:.6f}',
      sep='\n')
```

## 6 Тестирование модели

Сформируем эмббединги для тестовых изображений:

```python
df_test_filtered = df_test.drop_duplicates(subset='image')
test_img_resnet = func_img_embeddings(resnet, f'{PATH}/test_images', df_test_filtered)
test_img_resnet.shape
```

Создадим функцию для формирования эмбеддингов текстовых описаний тестовых изображений:

```python
def func_text_to_vector(text):
    #text = func_lemm_text(text)
    tokenized = tokenizer.encode(text, add_special_tokens=True, truncation=True)
    padded = np.array(tokenized + [0]*(max_len - len(tokenized)))
    attention_mask = np.where(padded != 0, 1, 0)
    with torch.no_grad():
        embeddings = model_bert(
        torch.LongTensor(padded).unsqueeze(0).to(device), 
        attention_mask=torch.LongTensor(attention_mask).unsqueeze(0).to(device)
      )
    return embeddings[0][:,0,:].cpu().numpy()[0]
```

Далее создадим функции для поиска изображений и их проверки на наличие запрещенного контента:

```python
def func_img_search(query, expected_image):
    if func_check_stoplist(query):
        print(f'Запрос:\n {query}')
        print(f'Результат:\n This image is unavailable in your country in compliance with local laws\n')
    
    else:
        text_vector = func_text_to_vector(query)
        full_vectors = np.concatenate((
            test_img_resnet, 
            np.vstack([text_vector] * test_img_resnet.shape[0])
          ), axis=1)
        target_predict = nn_model(full_vectors)
    
        imgs_path = df_test_filtered['image'].to_numpy()
        result = np.column_stack((imgs_path, target_predict))
        result = result[result[:, 1].argsort()]
        func_show_result(query, expected_image, result)
    
def func_check_stoplist(query):
    lemmas = lemmatizer(query)
    word_list = {word.lemma_ for word in lemmas}
    return bool(word_list & child_stop_dict)
```

Далее создадим функции для выведения изображений, релевантных запросу, на экран:

```python
def func_show_result(query, expected_image, result):
    fig = plt.figure(figsize=(20,10))
    func_add_plot(fig, expected_image, 'Оригинальное фото', 1)
    best_result = np.flipud(result[-3:, :])

    for i, (path, target) in enumerate(best_result, start=1):
        title = f'Топ {i} (Релевантность={float(target):.3f})'
        func_add_plot(fig, path, title, i + 1)

    plt.suptitle(f'Запрос:\n {query}')

def func_add_plot(figure, path, title, place):
    img = load_img(f'{PATH}/test_images/{path}')
    img_array = img_to_array(img)
    ax = figure.add_subplot(5, 5, place)
    ax.title.set_text(title)
    plt.imshow(img_array / 255.)
    #plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
```
